```python
"""
### ESCAPE (FROM THE SECRET TRANSYLVANIAN AIRPORT) ###

The adventure is inspired by Rocky Horror Picture Show.

Brad fly the Gulfstream jet out of hangar 3 to escape
from this Transylvanian airport. To get to hangar 3, he
needs to go through the first two, without getting near
either the runway or control tower.

The first two aircraft appeared are not right choices.

Brad can always use the magic word to escape.
"""


from time import sleep
import sys



# defining a function to handle unexpected entries
def tm_lost():
    tm = ("""
I'm terribly sorry. 
Your entry was lost in transmission. 
Please try again.
    """)
    for x in tm:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.005)

# defining a function for Brad to lose and replay the game
def fail():
    restart_or_not = input('<You\'ve failed the game. Do you want to restart? Y/N>\n')
    
    # I wanted to define an object as "loc" to identify where to restart, but failed
    # Tried to define loc=0 when exit at gate, loc=1 when exit at hangar_1, etc.
    # Tried to nest more conditions as below
    if restart_or_not == 'Y' or restart_or_not == 'y':
        # if loc == 0:
            # gate()
        # elif loc == 1:
            # hangar_1()
        # ... etc
            # ... etc
        # else:
        gate()
        
    elif restart_or_not == 'N' or restart_or_not == 'n':
        print('\n\nThank you for playing the game. Live long and prosper!')
        exit()

# defining a function for the part of the story where Brad dies
def death():
    do_you_copy = ("""
\nJanet:
- DO YOU COPY, CAPTAIN BRAD? COME IN!!!

- I'm so sorry, Captain Brad. 

- MAY YOU REST IN PEACE.\n""")
    for x in do_you_copy:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.01)
    sleep(1)
    print("""
              __)(__
        _____/      \\_____
        |  _     ___   _   ||
        | | \     |   | \  ||
        | |  |    |   |  | ||
        | |_/     |   |_/  ||
        | | \     |   |    ||
        | |   \. _|_. | .  ||
        |                  ||
        |    Brad Majors   ||
        |     ? - 2019     ||
        |                  ||
*       | *   **    * **   |**      **
\)),,..,/.,(//,,..,,\||(,,.,\\,.((//..
    """)
    fail()

# defining a function for the scenario when Brad hits the runway
def runway():
    runway_spotted = ("""
Janet:
- Oh, no!!! That's the runway! Run away from it!

- Captain Brad? Do you copy?
    """)
    for x in runway_spotted:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.01)
    death()

# defining a function for the scenario when Brad hits the control tower 
def control_tower():
    ct_spotted = ("""
Janet:
- That's the WELL-GUARDED control tower. Get away from it!

- S***! I see a group of people moving towards you!

- Do you copy, Captain Brad?
    """)
    for x in ct_spotted:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.01)
    death()

# defining a function for the scenario when Brad crashes a Boeing 747
def crash():
    print("""
\n<You successfully boarded the aircraft and managed to taxi to the runway.>
            """)
    sleep(0.5)
    print("""
Janet:
- Congratulations! This is such a successful rescue mission. 
            """)
    sleep(0.5)
    print("""
<The alarm just went off in the cockpit.>
            """)
    sleep(0.5)
    print("""
<You realized that the runway is not long enough to take off.>
            """)
    sleep(0.5)
    print("""
<Two seconds later, you realized it's too late to abort.>
            """)
    sleep(0.5)
    print("""
<The 747-8 jumbo jet crashed into the woods ahead of the runway.>
            """)
    death()

# defining the outcome when Brad survives
def win():
    print(f"\nCONGRATULATIONS! YOU'VE MADE IT!\n")
    print('*'*70)
    print('\x1b[1;37;43m' + """

 █████╗ ██╗    ██╗███████╗███████╗ ██████╗ ███╗   ███╗███████╗██╗██╗██╗
██╔══██╗██║    ██║██╔════╝██╔════╝██╔═══██╗████╗ ████║██╔════╝██║██║██║
███████║██║ █╗ ██║█████╗  ███████╗██║   ██║██╔████╔██║█████╗  ██║██║██║
██╔══██║██║███╗██║██╔══╝  ╚════██║██║   ██║██║╚██╔╝██║██╔══╝  ╚═╝╚═╝╚═╝
██║  ██║╚███╔███╔╝███████╗███████║╚██████╔╝██║ ╚═╝ ██║███████╗██╗██╗██╗
╚═╝  ╚═╝ ╚══╝╚══╝ ╚══════╝╚══════╝ ╚═════╝ ╚═╝     ╚═╝╚══════╝╚═╝╚═╝╚═╝

""" + '\x1b[0m')
    print('*'*70)
    exit()

# defining a function to help Rocky run the quiz
def quiz():
    
    # this is to count the remaining chances
    chance = 3
    while chance > 0:
        chance -= 1
        print("""
Rocky:
- Why don't you just shoot me?! 
But since you didn't, what's a bear without teeth?

- Be careful with your answer. I'm counting.
        """)
        answer = input('\n>>>')
        answer = str (answer)
        answer = answer.lower()
        
        # this is how to trigger the victory when given the right answer 
        if any(x in answer for x in ('shoot', 'kill', 'gummy')):
            print("You got it right!")
            win()
            break

        # this is what happens when the answer is wrong
        elif any(x not in answer for x in ('shoot', 'kill', 'gummy')):
            if chance > 0:
                print(f"Rocky:\n- Let's try again. You have {chance} guesses left.")
            elif chance == 0:
                print("Rocky:\n-You're out of guesses.\n<You suddenly got shot.>")
                death()
                break
            else:
                tm_lost()
        else:
            tm_lost()

# this is how the game starts
def start():
    print('#'*80)
    print('\x1b[6;30;41m' + """
 ▄▄▄▄▄▄▄▄▄▄▄ ▄▄▄▄▄▄▄▄▄▄▄ ▄▄▄▄▄▄▄▄▄▄▄ ▄▄▄▄▄▄▄▄▄▄▄ ▄▄▄▄▄▄▄▄▄▄▄ ▄▄▄▄▄▄▄▄▄▄▄ ▄ ▄ ▄ 
▐░░░░░░░░░░░▐░░░░░░░░░░░▐░░░░░░░░░░░▐░░░░░░░░░░░▐░░░░░░░░░░░▐░░░░░░░░░░░▐░▐░▐░▌
▐░█▀▀▀▀▀▀▀▀▀▐░█▀▀▀▀▀▀▀▀▀▐░█▀▀▀▀▀▀▀▀▀▐░█▀▀▀▀▀▀▀█░▐░█▀▀▀▀▀▀▀█░▐░█▀▀▀▀▀▀▀▀▀▐░▐░▐░▌
▐░▌         ▐░▌         ▐░▌         ▐░▌       ▐░▐░▌       ▐░▐░▌         ▐░▐░▐░▌
▐░█▄▄▄▄▄▄▄▄▄▐░█▄▄▄▄▄▄▄▄▄▐░▌         ▐░█▄▄▄▄▄▄▄█░▐░█▄▄▄▄▄▄▄█░▐░█▄▄▄▄▄▄▄▄▄▐░▐░▐░▌
▐░░░░░░░░░░░▐░░░░░░░░░░░▐░▌         ▐░░░░░░░░░░░▐░░░░░░░░░░░▐░░░░░░░░░░░▐░▐░▐░▌
▐░█▀▀▀▀▀▀▀▀▀ ▀▀▀▀▀▀▀▀▀█░▐░▌         ▐░█▀▀▀▀▀▀▀█░▐░█▀▀▀▀▀▀▀▀▀▐░█▀▀▀▀▀▀▀▀▀▐░▐░▐░▌
▐░▌                   ▐░▐░▌         ▐░▌       ▐░▐░▌         ▐░▌          ▀ ▀ ▀ 
▐░█▄▄▄▄▄▄▄▄▄ ▄▄▄▄▄▄▄▄▄█░▐░█▄▄▄▄▄▄▄▄▄▐░▌       ▐░▐░▌         ▐░█▄▄▄▄▄▄▄▄▄ ▄ ▄ ▄ 
▐░░░░░░░░░░░▐░░░░░░░░░░░▐░░░░░░░░░░░▐░▌       ▐░▐░▌         ▐░░░░░░░░░░░▐░▐░▐░▌
 ▀▀▀▀▀▀▀▀▀▀▀ ▀▀▀▀▀▀▀▀▀▀▀ ▀▀▀▀▀▀▀▀▀▀▀ ▀         ▀ ▀           ▀▀▀▀▀▀▀▀▀▀▀ ▀ ▀ ▀ 
""" + '\x1b[0m')                                                                                                                                                                               
    print('#'*80)
    greeting = ("""
Janet:
- Good evening, Captain Brad!

- I'm Agent Janet, assisting you on Dr. Scott's rescue mission.
    
- Congrats for making it to this SECRET AIRPORT in TRANSYLVANIA. 

- Your mission is to find the right aircraft and escape with three
privates and Dr. Scott.

- So, good luck and fly safe, Captain Brad!\n""")      
    for x in greeting:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.02)
    
    grt_input = input('<Press Enter to Continue>\n')
    
    # easter egg
    if any(x in grt_input for x in ('great scott', 'time warp')):
        print("\nLet the toilet paper roll you to the victory!")
        sleep(1)
        win()

    # normal result
    else:
        gate()


# starting point, first stage
def gate():
    gate_briefing = ("""
Janet:
- I'm having trouble interpreting the radar signals. 

- Visibility: 10 meters.

- According to my sreen, you are at the gate, Captain.
There are three roads ahead of you.

Dr. Scott:
- Brad, where should we go?

<1>. The road to the North.
<2>. The road to the West.
<3>. The road to the East.
    """)
    for x in gate_briefing:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.001)
    
    # the only expected loop is when the entry is unexpected
    while True:
        gate_choice = input('>>> ')
        gate_choice = str(gate_choice)
        gate_choice = gate_choice.lower()
        
        # easter egg
        if any(x in gate_choice for x in ('great scott', 'time warp')):
            print("\nLet the toilet paper roll you to the victory!")
            sleep(1)
            win()
            break
        
        # triggering the runway scenario
        elif 'north' in gate_choice or '1' in gate_choice: 
            runway()
            break
        
        # move on to the next stage 
        elif 'west' in gate_choice or '2' in gate_choice:
            hangar_1()
            break
        
        # triggering the control tower scenario       
        elif 'east' in gate_choice or '3' in gate_choice:
            control_tower()
            break
            
        # handling unexpected entries
        else:
            tm_lost()

# the second stage
def hangar_1():
    h1_briefing = ("""
Janet:
- Oh Captain Brad, you've reached the first hangar!

- Intelligence suggest there is a Cessna 172 aircraft. It should be
a four-seat aircraft.

- Do you want to take this aircraft or move on to the next hangar?

<1>. Take this Cessna.
<2>. Explore another hangar.
    """)
    for x in h1_briefing:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.001)
    
    # the only expected loop is when the entry is unexpected
    while True:
        h1_choice = input('>>> ')
        h1_choice = str(h1_choice)
        h1_choice = h1_choice.lower()
        
        # great, another easter egg
        if any(x in h1_choice for x in ('great scott', 'time warp')):
            print("\nLet the toilet paper roll you to the victory!")
            sleep(1)
            win()
            break

        # triggering a failing situation
        elif any(x in h1_choice for x in('1', 'take', 'cessna', 'this')): 
            print("""
Dr. Scott:
- There are only four seats on this aircraft, Brad. What now?

Private Magenta:
- Captain, am I seeing a gang of soldiers riding motorbikes?

Private Columbia:
- I think you're right, Magneta. Wait, they're coming at us!

Private Riff:
- AND THEY HAVE GUNS! AHHHHHHH....
            """)
            fail()
            break
        
        # moving on to the next stage
        elif any(x in h1_choice for x in('2', 'explore', 'another', 'hangar')):
            hangar_2()
            break
        
        # handling unexpected entries
        else:
            tm_lost()

# the third stage
def hangar_2():
    h2_briefing = ("""
Janet:
- Oh, Captain! This is the second hangar! Intelligence indicates
there is a Boeing 747-8 aircraft.

- Have you seen it?

Do you want to take this 747-8 or move on to the next hangar?

<1>. Fly this beautiful baby and call it a night.
<2>. Explore another hangar.
<3>. Ask Dr. Scott.
    """)
    for x in h2_briefing:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.001)
    
    # the only expected loop is when the entry is unexpected
    while True:
        h2_choice = input('>>> ')
        h2_choice = str(h2_choice)
        h2_choice = h2_choice.lower()
        
        # an easter egg, again
        if any(x in h2_choice for x in ('great scott', 'time warp')):
            print("\nLet the toilet paper roll you to the victory!")
            sleep(1)
            win()
            break

        # triggering the crash scenario
        elif any(x in h2_choice for x in 
               ('1', 'take', 'this', 'baby', 'fly', 'call', 'night', '747', '8' )):
            crash()
            break
        
        # moving on to the next stage
        elif any(x in h2_choice for x in 
                 ('2', 'explore', 'another', 'hangar', 'move')):
            hangar_3()
            break
            
        # triggering a small loop, Dr. Scott is not helping at all
        elif any(x in h2_choice for x in 
                 ('3', 'ask', 'scott', 'dr', 'doctor')):
            h2_choice_2 = input("""
Dr. Scott:
- This aircraft looks so fancy. Me likey.

- What do you want to do?

<1> Take this 747.
<2> Go to the next hangar.

>>> """)
            # Just another easter egg
            if any(x in h2_choice_2 for x in ('great scott', 'time warp')):
                print("\nLet the toilet paper roll you to the victory!")
                sleep(1)
                win()
                break
            
            # moving on to the next stage
            elif any(x in h2_choice_2 for x in 
                 ('2', 'explore', 'another', 'next', 'hangar', 'move')):
                hangar_3()
                break
            
            # triggering the crash scenario again
            elif any(x in h2_choice_2 for x in 
               ('1', 'take', 'this', 'baby','fly', 'call', 'night', '747', '8' )):
                crash()
                break
                
            else:
                tm_lost()
        else:
            tm_lost()

# stage 4, finally
def hangar_3():
    h3_briefing = ("""
Rocky:
- At last, Captain Brad, you're here. As you can see, there's a
Gulfstream G550 private jet. It can take off on this short runway
and it can seat all of you.

- I'll let you fly away if you can answer my question correctly.
You have three chances.

Janet:
- It's a trap. Don't fall for it! 

Dr. Scott:
- What should we do? Do you:
<1>. Answer the question.
<2>. Shoot Rocky.
    """)
    for x in h3_briefing:
        print(x, end='')
        sys.stdout.flush()
        sleep(0.001)
    
    while True:
        h3_choice = input('>>>')
        h3_choice = str(h3_choice)
        h3_choice = h3_choice.lower()
        
        # as always, the same old easter egg
        if any(x in h3_choice for x in ('great scott', 'time warp')):
            print("\nLet the toilet paper roll you to the victory!")
            sleep(1)
            win()
            break
        
        # yeah~ win!
        elif any(x in h3_choice for x in ('2', 'shoot', 'kill', 'rocky')):
            print("""
<You killed Rocky and take over the aircraft.>

Janet:
- Oh, Brad! How brave you are!

<You took off successfully and no one's after you.>
            """)
            win()
            break
            
        # ughhhh, quizzzzzzzzz
        elif any(x in h3_choice for x in ('1', 'answer', 'question')):
            quiz()
            break
                
        else:
            tm_lost()


start()
            
            
```
